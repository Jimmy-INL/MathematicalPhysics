Fourier series

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.special import gamma
%matplotlib inline
```

```python
x = np.linspace(-2*np.pi,2*np.pi,1000)
```

```python
a0 = 1
b1 = -1/np.pi
b3 = -1/(3*np.pi)
b5 = -1/(5*np.pi)
```

```python
y = np.zeros_like(x)
plt.plot(x,y);
```

```python
y = y + (1/2)*a0
plt.plot(x,y)
```

```python
y = y + b1*np.sin(x)
plt.plot(x,y)
```

```python
y = y + b3*np.sin(3*x)
```

```python
plt.plot(x,y)
```

```python
y = y + b5*np.sin(5*x)
plt.plot(x,y)
```

```python
a0 = 1
b1 = -1/np.pi
b3 = -1/(3*np.pi)
b5 = -1/(5*np.pi)
y = np.zeros_like(x) + (1/2)*a0 + b1*np.sin(x) + b3*np.sin(3*x) + b5*np.sin(5*x)
plt.plot(x,y)
```

```python
a0 = 1
b1 = 1/np.pi
b3 = 1/(3*np.pi)
b5 = 1/(5*np.pi)
y = np.zeros_like(x) + (1/2)*a0 + b1*np.sin(x) + b3*np.sin(3*x) + b5*np.sin(5*x)
plt.plot(x,y)
```

```python
from ipywidgets import interact
```

```python
def plotn(N):
    y = np.zeros_like(x)
    for n in range(1,N):
        y = y + np.cos(n*x)
    plt.plot(x,y)
interact(plotn,N=(1,1000))
```
