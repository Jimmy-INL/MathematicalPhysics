```python
%pylab inline
from __future__ import division
```

```python
c0 = 1.0
a = 0.2
L = 1.0
D = 1.0
Nterms = 100
Npoints = 200
```

```python
x = linspace(-L,L,num=Npoints)
def c(x=x,t=0.1,N=Nterms):
    result = zeros_like(x)
    for n in range(1,Nterms+1):
        result = result + 2*sin(n*pi*a/L)/(n*pi)*exp(-t*D*n**2*pi**2/L**2)*cos(x*n*pi/L)
    return 1 - (a/L) - result
```

```python
plot(x,c(x,0.1))
```

```python
import matplotlib.pyplot as plt
from JSAnimation import IPython_display
from matplotlib import animation

fig = plt.figure()
ax = plt.axes(xlim=(-L,L),ylim=(0,c0))
line, = ax.plot([],[],lw=2,label='concentration')
plt.legend(fancybox=True)

def init():
    line.set_data([],[])
    plt.grid(True)
    return line,

def conc(t):
    line.set_data(x,c(t=t/500.))
    return line,

animation.FuncAnimation(fig,conc,init_func=init,frames=100)
```
