We have the function 

$$
f(x) = \left\{\begin{matrix}
3x^2 + 2x^3 & -1<x<0\\ 
3x^2 - 2x^3 &  0<x<1
\end{matrix}\right.
$$

which, as per the hand-written solution, has series expansion

$$
f(x) = \frac{1}{2} - \frac{48}{\pi^4}\sum_{n=1,\textrm{odd}}^\infty \frac{cos(n\pi x)}{n^4}
$$

Let's plot the function and the series

```python
from __future__ import division
%matplotlib inline
import numpy as np, matplotlib as npl, pandas as pd, scipy as sp, seaborn as sns
from matplotlib import pyplot as plt
```

```python
# Let's all agree to only operate on arrays
def f(x):
    return np.piecewise(x, [x<=0, x>0], [lambda x: 3*x**2 + 2*x**3, lambda x: 3*x**2 - 2*x**3])
x = np.linspace(-1,1,1000)
plt.plot(x,f(x))
```

```python
x = np.linspace(-2,2,1000)
def a(n):
    if n == 0: return 1
    else:
        if n%2 == 0: 
            return 0 # even
        else: 
            return (-48/np.pi**4) * (1/n**4)
def b(n):
    return 0
def fourier(x,n):
    result = np.zeros_like(x) + 0.5*a(0)
    for i in range(1,n+1):
        result = result + a(i)*np.cos(i*x) + b(i)*np.sin(i*x)
    return result
plt.plot(x,fourier(x,10000))
```
