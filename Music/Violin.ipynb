```python
from __future__ import division
import numpy as np
from numpy import pi, sin, cos, array, linspace, zeros_like
import matplotlib.pyplot as plt
from matplotlib.pyplot import plot
%matplotlib inline
from JSAnimation import IPython_display
from matplotlib import animation
```

```python
c = 1.0 # wave speed
l = 2.0 # distance from nut to bridge
x0 = 0.21 # bow location
x0_array = array((x0,))
alpha = x0/c
A = 1.0 # maximum amplitude at x0

Npoints = 100
x = linspace(0,l,Npoints)
```

```python
def b(n):
    return ((2*A*l*l)/(n*n*pi*pi*c*alpha*(l-c*alpha))) * sin(n*pi*c*alpha/l)
```

```python
ns = array(range(1,11))
plot(ns,b(ns),'bo')
```

```python
def y(t,N,x=x):
    result = zeros_like(x)
    for n in range(1,N+1):
        result = result + b(n)*(sin(n*pi*x/l)*cos(n*pi*c*t/l)) / (sin(n*pi*x0/l))
    return result
```

```python
plot(x,y(0,100))
```

```python
fig = plt.figure()
ax = plt.axes(xlim=(0, l), ylim=(-3, 3))
line, = ax.plot([], [], lw=2, color='blue')
point, = ax.plot([], [], lw=5, color='black', marker='o')

# initialization function: plot the background of each frame
def init():
    line.set_data([], [])
    point.set_data([], [])
    return line,point

def animate(t,N=100):
    line.set_data(x,y(t/10.0,N))
    point.set_data(x0_array,y(t/10.0,N,x=x0_array))
    return line,point

animation.FuncAnimation(fig, animate, init_func=init,
                               frames=100,)
```
